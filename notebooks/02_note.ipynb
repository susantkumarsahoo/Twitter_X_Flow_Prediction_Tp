```python
import pandas as pd

def apply_pivot_examples(dataset_path: str) -> pd.DataFrame:
    # Load dataset
    df = pd.read_excel(dataset_path)

    # Select required columns (optional: ensures only relevant columns are used)
    required_columns = ['COMPLAINT TYPE', 'DEPT', 'CLOSED/OPEN']
    df = df[required_columns].copy()

    # Create pivot table: count of complaints grouped by COMPLAINT TYPE, DEPT, and CLOSED/OPEN
    pivot_table_df = df.pivot_table(
        index='COMPLAINT TYPE',
        columns=['DEPT', 'CLOSED/OPEN'],
        values='CLOSED/OPEN',  # counting occurrences
        aggfunc='count',
        fill_value=0
    )

    # Flatten multi-index columns for readability
    pivot_table_df.columns = [
        f"{dept}_{status}" for dept, status in pivot_table_df.columns
    ]

    # Add total columns per department and grand total
    dept_names = df['DEPT'].dropna().unique()
    for dept in dept_names:
        closed_col = f"{dept}_CLOSED"
        open_col = f"{dept}_OPEN"
        if closed_col in pivot_table_df.columns and open_col in pivot_table_df.columns:
            pivot_table_df[f"{dept}_TOTAL"] = pivot_table_df[closed_col] + pivot_table_df[open_col]
        elif closed_col in pivot_table_df.columns:
            pivot_table_df[f"{dept}_TOTAL"] = pivot_table_df[closed_col]
        elif open_col in pivot_table_df.columns:
            pivot_table_df[f"{dept}_TOTAL"] = pivot_table_df[open_col]

    # Add grand total column
    total_cols = [col for col in pivot_table_df.columns if col.endswith('_TOTAL')]
    pivot_table_df['GRAND_TOTAL'] = pivot_table_df[total_cols].sum(axis=1)

    # Reset index if needed
    pivot_table_df = pivot_table_df.reset_index()

    return pivot_table_df
```

```python
import pandas as pd
from typing import List, Dict, Optional

def apply_pivot_examples(dataset_path: str) -> pd.DataFrame:
    # Load dataset
    df = pd.read_excel(dataset_path)

    # Ensure required columns exist
    required = ['COMPLAINT TYPE', 'DEPT', 'CLOSED/OPEN']
    missing = [c for c in required if c not in df.columns]
    if missing:
        raise ValueError(f"Missing required columns: {missing}")

    # Work on a copy and clean text
    d = df[required].copy()

    # Normalize strings: strip, lower, then title-case
    def norm(s: Optional[str]) -> str:
        if pd.isna(s):
            return '(blank)'
        s = str(s).strip()
        return '(blank)' if s == '' else s

    d['COMPLAINT TYPE'] = d['COMPLAINT TYPE'].map(norm)
    d['DEPT'] = d['DEPT'].map(norm)
    d['CLOSED/OPEN'] = d['CLOSED/OPEN'].map(norm)

    # Optional: unify common variants in complaint types
    # Map variations to a single canonical form
    type_map: Dict[str, str] = {
        'Civil Works': 'Civil works',
        'NO Power Supply': 'No Power Supply',
        'Pole Shifting / Lt Sagging': 'Pole Shifting / Lt Sagging',
        'Transformer Failure / NCC': 'Transformer failure / NCC',
    }
    d['COMPLAINT TYPE'] = d['COMPLAINT TYPE'].replace(type_map)

    # Normalize DEPT values to desired set
    dept_map: Dict[str, str] = {
        'Commercial': 'Commercial',
        'O&M': 'O&M',
        'Operation & Maintenance': 'O&M',
        'Other': 'Other',
        'Others': 'Other',
        '(blank)': '(blank)',
    }
    d['DEPT'] = d['DEPT'].map(lambda x: dept_map.get(x, x))

    # Normalize CLOSED/OPEN values to exactly CLOSED / OPEN
    status_map: Dict[str, str] = {
        'Closed': 'CLOSED',
        'closed': 'CLOSED',
        'OPEN': 'OPEN',
        'Open': 'OPEN',
        'open': 'OPEN',
        '(blank)': '(blank)',
    }
    d['CLOSED/OPEN'] = d['CLOSED/OPEN'].map(lambda x: status_map.get(x, x))

    # Build pivot with counts using size (do not set `values`)
    pivot = d.pivot_table(
        index='COMPLAINT TYPE',
        columns=['DEPT', 'CLOSED/OPEN'],
        aggfunc='size',
        fill_value=0
    )

    # Ensure consistent column order
    desired_depts: List[str] = ['Commercial', 'O&M', 'Other', '(blank)']
    desired_status: List[str] = ['CLOSED', 'OPEN']
    # Reindex columns to the desired multi-index order, keep missing combinations as 0
    pivot = pivot.reindex(
        pd.MultiIndex.from_product([desired_depts, desired_status], names=pivot.columns.names),
        axis=1,
        fill_value=0
    )

    # Add per-department TOTAL columns
    for dept in desired_depts:
        closed_col = (dept, 'CLOSED')
        open_col = (dept, 'OPEN')
        total_series = pivot.get(closed_col, 0) + pivot.get(open_col, 0)
        pivot[(dept, 'TOTAL')] = total_series

    # Add GRAND TOTAL across department TOTALs
    total_cols = [(dept, 'TOTAL') for dept in desired_depts]
    pivot[('Grand', 'TOTAL')] = pivot[total_cols].sum(axis=1)

    # Sort rows alphabetically (optional: or by Grand TOTAL descending)
    # pivot = pivot.sort_values(('Grand', 'TOTAL'), ascending=False)

    # Flatten columns to match your readability preference
    pivot.columns = [f"{lvl0} {lvl1}".strip() for lvl0, lvl1 in pivot.columns]

    # Reset index to have COMPLAINT TYPE as a column
    pivot = pivot.reset_index()

    return pivot
```

```python
import pandas as pd
import plotly.graph_objects as go

def visualize_line_complaints():
    # Define the data
    data = {
        "Complaint Type": [
            "Billing", "Civil works", "Electrification", "Low Voltage", "Metering",
            "No Power Supply", "NSC", "Others", "Payment", "Pole Shifting / Lt Sagging",
            "Portal Problem", "Power Outage", "Safety", "Solar", "Transformer failure / NCC", "(blank)"
        ],
        "Commercial_CLOSED": [805, 0, 0, 0, 192, 0, 746, 0, 138, 0, 249, 0, 0, 70, 0, 0],
        "Commercial_OPEN": [3, 0, 0, 0, 0, 0, 5, 0, 0, 0, 0, 0, 0, 0, 0, 0],
        "O&M_CLOSED": [11, 1729, 652, 1306, 1, 7615, 6, 382, 1, 972, 1, 10474, 1798, 0, 783, 1],
        "O&M_OPEN": [0, 143, 0, 1, 0, 22, 0, 4378, 0, 71, 4, 39, 68, 0, 2, 0],
        "Other_CLOSED": [5, 5, 0, 0, 0, 22, 3, 11, 0, 1, 0, 39, 4, 0, 1, 0],
        "Other_OPEN": [0]*16,
        "Blank_CLOSED": [0]*16,
        "Blank_OPEN": [0]*16
    }

    df = pd.DataFrame(data)

    # Define department-status combinations and colors
    columns = [
        ("Commercial_CLOSED", "#1f77b4"),
        ("Commercial_OPEN", "#17becf"),
        ("O&M_CLOSED", "#2ca02c"),
        ("O&M_OPEN", "#98df8a"),
        ("Other_CLOSED", "#ff7f0e"),
        ("Other_OPEN", "#ffbb78"),
        ("Blank_CLOSED", "#9467bd"),
        ("Blank_OPEN", "#c5b0d5")
    ]

    # Create line traces
    traces = []
    for col, color in columns:
        traces.append(go.Scatter(
            x=df["Complaint Type"],
            y=df[col],
            mode='lines+markers',
            name=col.replace("_", " "),
            line=dict(color=color, width=2),
            marker=dict(size=6)
        ))

    # Build the figure
    fig = go.Figure(data=traces)
    fig.update_layout(
        title='Line Plot: CLOSED and OPEN Complaints by Department and Complaint Type',
        xaxis_title='Complaint Type',
        yaxis_title='Number of Complaints',
        legend_title='Department and Status',
        template='plotly_white',
        height=650
    )

    # Show the chart
    fig.show()

# Call the function
visualize_line_complaints()
```
